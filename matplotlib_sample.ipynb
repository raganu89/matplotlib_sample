```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,'r-')
```

```python
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(x,y,'b')
```

```python
fig = plt.figure()
axes = fig.add_axes([.1,.1,.8,.8])
axes.plot(x,y)
```

```python
fig = plt.figure()
axes1 = fig.add_axes([.1,.1,.8,.8])# left bottom width and hegiht
axes2 = fig.add_axes([.2,.5,.4,.3])
```

```python
fig = plt.figure()
axes1 = fig.add_axes([.1,.1,.8,.8])# left bottom width and hegiht
axes2 = fig.add_axes([.2,.5,.4,.3])
axes1.plot(x,y,'r')
axes2.plot(y,x)
```

```python
fig = plt.figure()
axes = fig.add_axes([.1,.1,.8,.8])
axes.plot(x,y,color='purple',lw=3,linestyle=':')
```

```python
fig = plt.figure()
axes = fig.add_axes([.1,.1,.8,.8])
axes.plot(x,y,color='purple',lw=3,ls='solid')
```

```python
fig = plt.figure()
axes = fig.add_axes([.1,.1,.8,.8])
axes.plot(x,y,color='purple',lw=3,ls='solid',marker = 'o')
```
